```python
# Importing libraries
import sqlite3 
import pandas as pd
```

```python
#Creating SQLite connection
conn = sqlite3.connect("../sql/retail.db")
```

```python
# Loading data into SQLite
df.to_sql("retail_sales", conn, if_exists="replace", index=False)
conn.close()
```
